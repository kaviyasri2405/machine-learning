```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
# Generate a synthetic dataset for Bank Loan prediction
np.random.seed(42)

num_samples = 1000

data = {
    'ApplicantIncome': np.random.normal(50000, 15000, num_samples).astype(int),
    'CoapplicantIncome': np.random.normal(15000, 10000, num_samples).astype(int).clip(min=0),
    'LoanAmount': np.random.normal(150000, 50000, num_samples).astype(int).clip(min=10000),
    'CreditHistory': np.random.choice([0, 1], num_samples, p=[0.2, 0.8]),
    'Gender': np.random.choice(['Male', 'Female'], num_samples),
    'Married': np.random.choice(['Yes', 'No'], num_samples),
    'Education': np.random.choice(['Graduate', 'Not Graduate'], num_samples),
    'SelfEmployed': np.random.choice(['Yes', 'No'], num_samples, p=[0.15, 0.85]),
    'Loan_Status': np.random.choice(['Y', 'N'], num_samples, p=[0.7, 0.3]) # Target variable
}

df = pd.DataFrame(data)

# Introduce some missing values for demonstration
for col in ['Gender', 'Married', 'SelfEmployed', 'LoanAmount']:
    df.loc[df.sample(frac=0.05).index, col] = np.nan

print("First 5 rows of the dataset:")
display(df.head())
```

```python
# Preprocessing: Handle missing values
# For numerical features, use median imputation
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
df['CoapplicantIncome'] = df['CoapplicantIncome'].fillna(df['CoapplicantIncome'].median())
df['ApplicantIncome'] = df['ApplicantIncome'].fillna(df['ApplicantIncome'].median())

# For categorical features, use mode imputation
for col in ['Gender', 'Married', 'SelfEmployed']:
    df[col] = df[col].fillna(df[col].mode()[0])

# Encode categorical features
le = LabelEncoder()
for col in ['Gender', 'Married', 'Education', 'SelfEmployed']:
    df[col] = le.fit_transform(df[col])

# Convert target variable 'Loan_Status' to numerical (0 and 1)
df['Loan_Status'] = df['Loan_Status'].map({'Y': 1, 'N': 0})

print("Dataset after preprocessing:")
display(df.head())
```

```python
# Define features (X) and target (y)
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# Initialize and train the Gaussian Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

print("Naive Bayes model trained successfully.")
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```
