```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset as an example
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the K-NN classifier
k = 3  # You can choose any odd number for k to avoid ties
knn_classifier = KNeighborsClassifier(n_neighbors=k)

# Train the model
knn_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"K-Nearest Neighbors (k={k}) Classifier Accuracy: {accuracy:.2f}")

# Example of making a single prediction
new_data = np.array([[5.1, 3.5, 1.4, 0.2]])  # Example new data point (e.g., a new iris flower)
prediction = knn_classifier.predict(new_data)
print(f"Prediction for new data {new_data[0]}: {iris.target_names[prediction[0]]}")
```
