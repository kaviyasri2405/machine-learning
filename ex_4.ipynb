```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
class NeuralNetwork:
    def __init__(self, layers):
        self.layers = layers
        self.weights = []
        self.biases = []

        for i in range(len(layers) - 1):
            self.weights.append(np.random.rand(layers[i], layers[i+1]))
            self.biases.append(np.random.rand(1, layers[i+1]))

    def forward_propagation(self, inputs):
        self.activations = [inputs]
        for i in range(len(self.weights)):
            net_input = np.dot(self.activations[-1], self.weights[i]) + self.biases[i]
            output = sigmoid(net_input)
            self.activations.append(output)
        return self.activations[-1]

    def backward_propagation(self, inputs, targets, learning_rate):
        output_error = targets - self.activations[-1]
        output_delta = output_error * sigmoid_derivative(self.activations[-1])

        deltas = [output_delta]

        for i in range(len(self.weights) - 1, 0, -1):
            error = deltas[0].dot(self.weights[i].T)
            delta = error * sigmoid_derivative(self.activations[i])
            deltas.insert(0, delta)

        for i in range(len(self.weights)):
            self.weights[i] += self.activations[i].T.dot(deltas[i]) * learning_rate
            self.biases[i] += np.sum(deltas[i], axis=0, keepdims=True) * learning_rate

    def train(self, X, y, epochs, learning_rate):
        for epoch in range(epochs):
            for i in range(len(X)):
                input_data = np.array([X[i]])
                target_data = np.array([y[i]])

                self.forward_propagation(input_data)
                self.backward_propagation(input_data, target_data, learning_rate)

            if epoch % 1000 == 0:
                loss = np.mean(np.square(y - self.forward_propagation(X)))
                print(f"Epoch {epoch}, Loss: {loss:.4f}")

    def predict(self, inputs):
        return self.forward_propagation(inputs)
```

```python
# Example Usage: XOR problem
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Define the network architecture (2 input neurons, 4 hidden neurons, 1 output neuron)
neural_network = NeuralNetwork(layers=[2, 4, 1])

# Train the network
neural_network.train(X, y, epochs=10000, learning_rate=0.1)

# Test the trained network
print("\nPredictions after training:")
for i in range(len(X)):
    prediction = neural_network.predict(np.array([X[i]]))
    print(f"Input: {X[i]}, Expected: {y[i]}, Predicted: {prediction[0][0]:.4f}")
```
